```python
# Statistical Analysis

# Summary Statistics
df.describe()
```

```python
# Average Score of Each Subject

subject_means = df[["math score", "reading score", "writing score"]].mean()
print(subject_means)
```

```python
# Highest Average Subject
highest_avg_subject = subject_means.idxmax()
print("Highest Average Subject:", highest_avg_subject)

# Lowest Average Subject
lowest_avg_subject = subject_means.idxmin()
print("Lowest Average Subject:", lowest_avg_subject)
```

```python
# Highest Score in Each Subject
print("Highest Score in Each Sub:")
print(df[["math score", "reading score", "writing score"]].max())

# Lowest Score in Each Subject
print("\nLowest Score in Each Sub:")
print(df[["math score", "reading score", "writing score"]].min())
```

## Observations

- The average Reading score (69.17) is the highest among all subjects.
- The average Mathematics score (66.09) is the lowest among all subjects.
- Students generally perform better in Reading and Writing compared to Mathematics.
- Perfect scores (100) were achieved in all three subjects, indicating the presence of high-performing students.
- The minimum Mathematics score is 0, while the minimum Reading and Writing scores are 17 and 10 respectively, showing significant variation in student performance.

```python
df["average score"] = (
    df["math score"] + 
    df["reading score"]+
    df["writing score"]
) / 3

df
```

```python
# Performance Analysis
df2 = df[["gender", "math score", "reading score", "writing score", "average score"]]

# Top 10 students
top_ten = df2.sort_values("average score", ascending=False).head(10)


# Bottom 10 students
bottom_ten = df2.sort_values("average score").head(10)
```

```python
top_ten
```

```python
bottom_ten
```

## Observations

- The top-performing students consistently achieved high scores across all three subjects.

- Several top-performing students obtained scores close to or equal to 100 in multiple subjects.

- The bottom-performing students showed low performance across multiple subjects rather than a single subject.

- Average score provides a balanced measure for evaluating overall student performance.

```python
# Gender-Based Analysis

df.groupby("gender")[
    ["math score", 
     "reading score", 
     "writing score", 
     "average score"
    ]
].mean()
```

## Observations

* Female students achieved a higher overall average score (69.57) compared to male students (65.84).

* Female students performed significantly better in Reading and Writing, with average scores above 72 in both subjects.

* Male students performed better in Mathematics, achieving an average score of 68.73 compared to 63.63 for female students.

* Despite stronger Mathematics performance among male students, female students demonstrated better overall academic performance due to higher Reading and Writing scores.

* The results suggest that performance differences exist across subjects between male and female students.

```python
# Test Preparation Analysis

df.groupby("test preparation course") [
    ["math score", 
     "reading score", 
     "writing score", 
     "average score"
    ]
].mean()
```

## Observations

* Students who completed the test preparation course achieved higher scores across all subjects.

* The average overall score of students who completed the course was 72.67, compared to 65.04 for students who did not complete it.

* Students who completed the course scored approximately 7–10 points higher on average in each subject.

* Reading and Writing showed particularly strong improvements among students who completed the preparation course.

* The results indicate that test preparation has a positive impact on student academic performance.

```python
# lunch Type Analysis

df.groupby("lunch") [
    ["math score", 
     "reading score", 
     "writing score", 
     "average score"
    ]
].mean()
```

## Observations

* Students with standard lunch achieved significantly higher scores across all subjects compared to students receiving free/reduced lunch.

* The average overall score of students with standard lunch was 70.84, while students with free/reduced lunch achieved an average score of 62.20.

* The performance gap is approximately 8.6 points in overall average score.

* Students with standard lunch consistently outperformed students with free/reduced lunch in Mathematics, Reading, and Writing.

* The results suggest that lunch type is associated with differences in academic performance.

```python
# Parental Education Analysis

df.groupby("parental level of education") [
    ["math score", 
     "reading score", 
     "writing score", 
     "average score"
    ]
].mean()
```

## Observations

* Students whose parents hold a master's degree achieved the highest overall average score (73.60).

* Students whose parents hold a bachelor's degree also demonstrated strong academic performance, with an average score of 71.92.

* Students whose parents completed only high school recorded the lowest overall average score (63.10).

* A general upward trend can be observed between parental education level and student academic performance.

* Students with parents holding higher educational qualifications tend to achieve better scores across Mathematics, Reading, and Writing.
